2. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.
```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')

# Opening JSON file
file = open('songLyrics.json')
 
# load JSON object 
data = json.load(file)

# Get Lyrics
lyrics = data['GetLyricResult']['Lyric']
print(lyrics)

#Get Analysis
doc = nlp(lyrics)
print(f'\nPolarity Score: {doc._.blob.polarity}')

#The resulting polarity score of approximately -0.13 indicates a more negative connotiation of the lyrics. 
#While the score is negative, it's fairly close to 0 (neutral connotation), which is interesting given the content of the lyrics. I would have scored more negative based on words like "evil", "killer", etc.
```
